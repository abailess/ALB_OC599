```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data points for the first set of lines (slope = -1.5 and -3)
x1 = np.linspace(0, 10, 100)
y1 = -1 * (x1 - 5)  # Adjusted intercept to cross at x=5
y2 = -1.75 * (x1 - 5)    # Adjusted intercept to cross at x=5

# Generate data points for the second set of lines (slope < original, lower intercept)
x2 = np.linspace(0, 10, 100)
y3 = -2.25 * (x2 - 5) - 10  # Larger downward shift in intercept
y4 = -3 * (x2 - 5) - 10  # Larger downward shift in intercept

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the first set of lines (darker colors)
plt.plot(x1, y1, color='b', linewidth=2)
plt.plot(x1, y2, color='g', linewidth=2)

# Plot the second set of lines (lighter colors)
plt.plot(x2, y3, color='lightblue', linewidth=2)
plt.plot(x2, y4, color='lightgreen', linewidth=2)

# Set the x and y axis labels
plt.xlabel('Log size', fontsize='18')
plt.ylabel('Log abundance', fontsize='18')

# Remove axis ticks and labels
plt.xticks([])
plt.yticks([])

# Set the axis limits to center the lines
plt.xlim(0, 10)
plt.ylim(-20, 10)  # Adjusted downward shift

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data for a stochastic line
x = np.linspace(0, 10, 100)
y = 0.5 + np.random.normal(0, 0.2, 100)

# Create irregularly spaced points on the line
x_points = np.array([1, 2.5, 4.3, 5.2, 6.8, 8.0])
y_points = 0.5 + np.random.normal(0, 0.1, 6)

# Calculate error bars to touch the line
y_error = np.abs(0.5 - y_points) + 0.02  # Adjusted error bars

# Create the figure and axes
fig, ax = plt.subplots(figsize=(8, 4))

# Plot the stochastic line in a shade of blue
ax.plot(x, y, label='Stochastic Line', color='#4E79A7')

# Plot the irregularly spaced points in a matching shade of green with thicker lines and error bars
ax.errorbar(x_points, y_points, yerr=y_error, fmt='o', label='Points with Error Bars', color='#59A14F', linewidth=2, elinewidth=2)

# Add labels and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data for a stochastic line
x = np.linspace(0, 10, 100)
y = 0.5 + np.random.normal(0, 0.2, 100)

# Create irregularly spaced points on the line
x_points = np.array([1, 2.5, 4.3, 5.2, 6.8, 8.0])
y_points = 0.5 + np.random.normal(0, 0.1, 6)

# Calculate error bars to touch the line
y_error = np.abs(0.5 - y_points) + 0.02  # Adjusted error bars

# Create the figure and axes
fig, ax = plt.subplots(figsize=(8, 4))

# Plot the stochastic line as a dashed blue line
ax.plot(x, y, label='Stochastic Line', color='#4E79A7', linestyle='--')

# Plot the irregularly spaced points in a more prominent shade of green with thicker lines and error bars
ax.errorbar(x_points, y_points, yerr=y_error, fmt='o', label='Points with Error Bars', color='#59A14F', linewidth=2, elinewidth=2)

# Add labels and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Set a seed for reproducibility
#seed = 12
#np.random.seed(seed)

# Create data for a stochastic line
x = np.linspace(0, 10, 100)
y = 0.5 + np.random.normal(0, 0.02, 100)  # Smaller noise

# Create irregularly spaced points on the line
x_points = np.array([1, 2.5, 4.3, 5.2, 6.8, 8.0])
y_points = 0.5 + np.random.normal(0, 0.01, 6)  # Smaller error range

# Calculate error bars to touch the line (order of magnitude smaller)
y_error = np.abs(0.5 - y_points) + 0.002  # Adjusted error bars

# Perform linear regression to calculate the p-value
slope, intercept, r_value, p_value, std_err = stats.linregress(x_points, y_points)

# Calculate the correlation coefficient (r-value) between the blue line and the green points
r_value, _ = stats.pearsonr(y[:len(x_points)], y_points)

# Create the figure and axes
fig, ax = plt.subplots(figsize=(8, 4))

# Plot the stochastic line as a dashed blue line
ax.plot(x, y, label='Stochastic Line', color='#4E79A7', linestyle='--')

# Plot the irregularly spaced points in a more prominent shade of green with thicker lines and error bars
ax.errorbar(x_points, y_points, yerr=y_error, fmt='o', label='Points with Error Bars', color='#59A14F', linewidth=2, elinewidth=2)

# Add labels and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add the correlation coefficient (r-value) above the legend
ax.text(0.65, 0.9, f'Correlation (r) = {r_value:.4f}', fontsize=12, color='black', transform=ax.transAxes)


# Remove axis ticks for x and y
ax.set_xticks([])
ax.set_yticks([])

ax.legend()

# Show the plot
plt.show()
```
