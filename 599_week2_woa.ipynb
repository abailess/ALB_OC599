```python
#Dependencies
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cftime
```

```python
#Read WOA using opendap
temp_url = (
    "http://apdrc.soest.hawaii.edu:80/dods/public_data/WOA/WOA18/5_deg/annual/temp"
)
# Sal_url = (
#     "url"
# )
```

```python
# Use xarray open dataset at url (tmn is the variable)
dstemp = xr.open_dataset(temp_url).tmn.load()
```

```python
#dstemp.sel(lev=5).plot() #this will plot the values based on meters or the VALUES (0m)

#dstemp.isel(lev=0).plot() #this will plot the values based on the level it is indexed (0 = 0m, 1 = 5m)

dstemp.sel(lev=7, method ='nearest').plot() #this will plot the values at 7m, if there is no 7m, it will plot the depth that is closest to 7m, in this case 5m
```
